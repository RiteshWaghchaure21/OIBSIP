```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(df)
```

```python
kmeans = KMeans(n_clusters=8, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
silhouette = silhouette_score(X, labels)
print("Silhouette score for K-means:", silhouette)
```

```python
hier = AgglomerativeClustering(n_clusters=8, linkage="ward")
hier.fit(X)
labels = hier.labels_
silhouette = silhouette_score(X, labels)
print("Silhouette score for hierarchical:", silhouette)
```

```python
unique_labels = np.unique(labels)
print("Number of unique labels:", len(unique_labels))
```

```python
labels = np.abs(labels)
```

```python
labels = labels[labels >= 0]
```

```python
label_counts = {}
for label in labels:
    if label not in label_counts:
        label_counts[label] = 0
    label_counts[label] += 1
```

```python
negative_values = np.where(labels < 0)[0]
```

```python
labels[negative_values] = 0
labels = labels[labels >= 0]
```

```python
label_counts = np.bincount(labels)
```

```python
print(label_counts)
```

```python
print("Label counts:", label_counts)
```
